#Data is loaded

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

imported pandas as pd and numpy as np. Imported matplot library too. 

```python
path = "/content/forbesathletesv2.csv"
```

For some reason i couldnot make the link work. So, i downloaded the .csv file and uploated on collab and then copied its path.

```python
df = pd.read_csv(path)
```

```python
df.head(10)
```

Top 10 heads of 2021. 

```python
year_group = df.sort_values(by='Year', ascending=False).groupby('Year')
```

Creating a varoable to group the data based on the year.

```python
each_year = year_group['Earnings'].sum() 
list_year = df.Year.unique() 
list_year = sorted(list_year, reverse=False) 
```

Creating the total of each year total income. 

```python
plt.figure(figsize=(13,6))
plt.bar(list_year, each_year, color='lightblue')

plt.title('Total Athletes Income per Year', fontsize=18)
plt.ylabel('Million USD', fontsize=16, color='green')
plt.yticks(fontsize=14)
plt.xticks(fontsize=14)

plt.grid(linestyle='--', axis='y')
```

Visualization of the data

```python
name_group = df.groupby('Name') #Count the total earnings of every athlete
total_earning = [i for i in name_group['Earnings'].sum()]

names = df.Name.unique() #Create a list of names
names.sort() #Sort it to make it corresponds to the total_earning list

all_data = {'Name':names, 'TotalEarnings':total_earning}
```

Counting the total earnings of every athelete.

```python
df2 = pd.DataFrame(all_data)
df2.sort_values('TotalEarnings', ascending=False, inplace=True)
```

Creating a new data frame df2

```python
df2.head()
```

```python
top10_names = df2['Name'].head(10)
top10_TEs = df2['TotalEarnings'].head(10)
```

Creating a new variable for the plot.

```python
plt.figure(figsize=(13,6))
plt.barh(top10_names, top10_TEs, edgecolor='black', color='yellow')

plt.title('Total Income', fontsize=18)
plt.xlabel('Million USD', color='green', fontsize=18)
plt.yticks(fontsize=14)
plt.xticks(fontsize=14)

plt.grid(linestyle='--', axis='x')
plt.gca().invert_yaxis()
```
